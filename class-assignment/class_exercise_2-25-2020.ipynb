#### Reading in the Bees File
Taking author counts and attributing praise to the best author

```python
import csv
import re

authorlist =[]
inputpath = "bee_list.txt"
with open(inputpath, "r") as dset:
    data = csv.DictReader(dset, delimiter='\t')
    header = data.fieldnames # Use this because of jupyter I/O file close
    for row in data:
        authorlist.append(row['Taxon Author'])
```

```python
author = str(authorlist).replace("(","")
author1 = str(author).replace(")","")
author2 = str(author1).replace(" &",",")
author3 = str(author2).replace("'","")
author4 = re.sub(r'[\d\s]','', author3)
author5 = str(author4).split(',')


from collections import defaultdict
count = defaultdict(int)
for a in author5:
    count[a] += 1
count
```
