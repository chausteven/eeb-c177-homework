# Part 1 (Figures)
-----

```python
import numpy as np
```

```python
inflam_data = numpy.loadtxt(fname="inflammation-01.csv", delimiter=',')
print(inflam_data)
```

```python
patientstd=np.std(inflam_data, axis=0)
print(patientstd)

patientmax=np.max(inflam_data, axis=0)
print(patientmax)

patientmin=np.min(inflam_data, axis=0)
print(patientmin)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(patientstd)
plt.xlabel('Days')
plt.ylabel('Standard Deviation between patient\'s inflammation score')
plt.title('Standard Deviation of Inflammation Score Each Day')
```

```python
plt.plot(patientmax)
plt.xlabel('Days')
plt.ylabel('Max inflammation score between patient\'s')
plt.title('Max Inflammation Score for Each Day')
```

```python
plt.plot(patientmin)
plt.xlabel('Days')
plt.ylabel('Min inflammation score between patient\'s')
plt.title('Min Inflammation Score for Each Day')
```

```python
fig, (pstd, pmax, pmin) = plt.subplots(3)

pstd.plot(patientstd)
pmax.plot(patientmax)
pmin.plot(patientmin)
```

```python
fig=plt.figure(figsize=(5,10))
plot1=fig.add_subplot(3,1,1)
plot2=fig.add_subplot(3,1,2)
plot3=fig.add_subplot(3,1,3)

fig.suptitle('Patients\' Inflammation Score', fontsize=20, fontweight='bold')

plot1.plot(patientstd)
plot1.set_ylabel('Standard Deviation')

plot2.plot(patientmax)
plot2.set_ylabel('Max')

plot3.plot(patientmin)
plot3.set_ylabel('Min')
plot3.set_xlabel('Days')
```

```python
#4
word = 'UCLA'
rev_word = ''
for switch in word:
    rev_word = switch + rev_word
    print(rev_word)
    
```

```python
#5
x = 5
coefs = [2, 4, 3]
y=0

for idx, val in enumerate(coefs):
    y += val * x**idx
    print(y)
```

# Part 3 (Lists)
---------

```python
#1 A list is a data structure that contains item(s) that retains their order and is mutable. 
```

```python
#2 To create a list you can either use the list() function or by using the [] and assigning it to a variable. 
```

```python
#3 You can change the value of an individual element by either offsetting using [] and stating the index you want to change and what you want to change it to, or you can splice the element [:] stating which indexs you would like to change and what you want to change it to.

#Offsetting
a = [1,2,3,4,5,6]
a[3]=100
print(a)

#Slicing
b = [1,2,3,4,5,6]
b[3:3]=[100]
print(b)
```

```python
#4 Two ways you can append a value to an existing list is by using the append() function or by using the insert() function to offset an item into the list.

#append()
x = [1,2,3,4,5]
x.append(6)
print(x)

#insert()
y = [1,2,3,4,5]
y.insert(5,6)
print(y)
```

```python
#5
string = 'hello'
letterlist=[]

for letter in string:
    letterlist.append(letter)
    print(letterlist)
```

# Part 4 (Conditionals)
-----

```python
#1
if 4 > 5: 
    print('A')
elif 4 == 5:
    print('B')
elif 4 < 5:
    print('C')
    
# This code ends up printing 'C'. Due to the 'if' statement the code first checks if 4 is greater than 5, which it is not so it goes to the next statement elif (else if). Next, the program checks if 4 is equal to 5, which it is not so it goes to the next statement. Finally, the program checks if 4 is smaller than 5, and at this point this is true so it runs the print() function to print out the string 'C'.
```

```python
#2
vowels = 'aeiouAEIOU'
phrase = 'I love to code, especially on python.'
vowel_list=[]

for letter in phrase :
    if letter in vowels:
        vowel_list += letter
print('There are', len(vowel_list), 'vowels in this string and they are', vowel_list, '.')
```

```python
#3
star = '* '
blank = ' '

for pyramid in range(6):
    space=6-pyramid
    print(blank*space, star*pyramid, blank*space)
    
```
